# 1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to keep all customers who have done transactions with us and select the join type accordingly.

```python
# renaming some coloumns
```

```python
Transactions.rename(columns={'prod_subcat_code':'prod_sub_cat_code'},inplace=True)
```

```python
Transactions.rename(columns={'cust_id':'customer_Id'},inplace=True)
```

```python
# mearging customer and transaction table 
```

```python
customer_1 = pd.merge( left= Customer,right = Transactions, left_on='customer_Id',right_on='customer_Id',how='right')
customer_1
```

```python
#final merging

customer_Final = pd.merge(left=customer_1,right=prod_cat_info,left_on=['prod_cat_code','prod_sub_cat_code'],
                          right_on = ['prod_cat_code','prod_sub_cat_code'],how = 'left')

customer_Final
```

```python
customer_Final.dtypes
```

```python
#datetime formating
customer_Final['DOB'] = customer_Final['DOB'].apply(lambda x : pd.to_datetime(x,format='%d-%m-%Y'))

customer_Final['tran_date'].replace('/','-')
customer_Final['tran_date'] = customer_Final['tran_date'].apply(lambda x : pd.to_datetime(x))
```

```python
# checking for duplicates
customer_Final.duplicated().sum()
```

```python
customer_Final.drop_duplicates(inplace=True)
```

```python
customer_Final.duplicated().sum()
```
